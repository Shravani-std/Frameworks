# Basic Math
Without Pytorch

```python
# 1.Vectors dot product
import numpy as np
a = np.array([1,2,3])
b = np.array([6,5,4])
dot = np.sum(a*b)
print("Dot Product: ", dot)

x = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([1, 1, 1])

print(x + b)
```

```python
# 2.Matrix Multiplication
def matrix_mul(A,B):
  n,m = A.shape
  m2,p = B.shape
  assert m == m2

  C = np.zeros((n,p))

  for i in range(n):
    for j in range(p):
      for k in range(m):
        C[i][j] += A[i][k] * B[k][j]
  return C

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(matrix_mul(A, B))
```

```python
# 3.Gradient Intuition
def numerical_grad(f,x,h=1e-5):
  return (f(x+h)-f(x-h) / (2*h))
def f(x):
  return x**2
x = 3.0
print("Numerical grad:", numerical_grad(f, x))
print("True grad:", 2*x)
```

```python
# Linear Regression from scartch
import numpy as np
np.random.seed(0)
X = np.random.randn(100,1)
true_w = 3
truw_b = 2

y = true_w*X + truw_b + 0.1*np.random.randn(100,1)

w = np.random.randn(1)
b = np.random.randn(1)
lr = 0.1

for epoch in range(1000):
  y_pred = w * X + b # Changed B to b
  loss = np.mean((y_pred - y)**2)

  dw = 2 * np.mean((y_pred - y) * X)
  db = 2 * np.mean(y_pred - y)

  w -= lr * dw
  b -= lr * db

print("Learned w:", w)
print("Learned b:", b)
```

# Pytorch Basics
Tensor Operations

```python
import torch

x = torch.tensor([[1.,2.], [3.,4.]])
print(x.shape)
print(x.dtype)
print(x.device)

y = torch.sin(x)
z = torch.matmul(x, x)
print(y)
print(z)
```

```python
x = torch.tensor(3.0, requires_grad=True)
y = x**2 + 2*x
y.backward()
print(x.grad)
```

**Linear Regression In Pytorch**

```python
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
lr = 0.01
for _ in range(1000):
  x = torch.randn(10)
  y = 3*x + 2
  y_pred = w*x +b
  loss = ((y_pred-y)**2).mean()

  loss.backward()
  with torch.no_grad():
    w-=lr*w.grad
    b-=lr*b.grad
  w.grad.zero_()
  b.grad.zero_()
```

**Neural Network**

```python
import torch.nn as nn

class MyNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(10,32)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(32,1)

  def forward(self, x):
    x = self.relu(self.fc1(x))
    return self.fc2(x)

model = MyNet()
model = torch.compile(model)

x = torch.randn(32,10)
output = model(x)
print(output)
```
